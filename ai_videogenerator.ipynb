```python
from diffusers import CogVideoXPipeline
import torch
```

```python
from diffusers.utils import export_to_video
```

```python
pipe = CogVideoXPipeline.from_pretrained(
    "THUDM/CogVideoX-2b",
    torch_dtype=torch.float16
)
```

```python
pipe.enable_sequential_cpu_offload()
pipe.vae.enable_tiling()
pipe.vae.enable_slicing()
```

```python
# Cell 4 — Generate each scene as a separate clip
prompts = {
    "scene1_before": (
        "Cluttered dark bedroom at night, gaming desk covered in cloth, "
        "cables on floor, bare dark parquet flooring, cold harsh lighting, "
        "cinematic wide shot from doorway, warm color grade"
    ),
    "scene2_timelapse": (
        "Time lapse hands laying light oak foam mats on dark parquet floor, "
        "Japanese floor mattress unfolding, forest green linen, "
        "aquarium being placed and filled, amber LED strip along skirting board, "
        "cinematic warm golden tones"
    ),
    "scene3_work_zone": (
        "Clean black desk dual monitors glowing softly, green pothos vines "
        "trailing behind screens, aquarium glowing blue-green below, "
        "amber LED gold reflection on oak mats, night city lights through window, "
        "cinematic shallow depth of field push-in"
    ),
    "scene4_sacred_zone": (
        "Prayer altar shelf, brass diyas flickering warm flame, framed Ganesh artwork, "
        "marigold petals, mala beads, copper kalash, incense smoke rising slowly, "
        "soft golden light, cinematic macro shot"
    ),
}

for scene_name, prompt in prompts.items():
    print(f"Generating {scene_name}...")
    video = pipe(
        prompt=prompt,
        num_frames=49,          # ~6 seconds at 8fps
        height=480,
        width=720,
        num_inference_steps=25,
        guidance_scale=6.0,
    ).frames[0]
    filename = f"{scene_name}.mp4"
    export_to_video(video, filename, fps=8)
    print(f"  ✅ Saved {filename}")
```

```python
# Cell 5 — Download all clips
from google.colab import files
import os

for f in ["scene2_timelapse.mp4",
          "scene3_work_zone.mp4", "scene4_sacred_zone.mp4"]:
          if os.path.exists(f):
            files.download(f)
```
